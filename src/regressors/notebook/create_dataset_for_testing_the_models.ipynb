```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')

from sklearn.datasets import load_boston
```

```python
boston_dataset = load_boston()
```

```python
feature_data = boston_dataset.data
target = boston_dataset.target
features_name = boston_dataset.feature_names
```

```python
df_boston = pd.DataFrame(feature_data, columns=features_name)
df_boston.head()
```

```python
df_boston['target'] = target
df_boston.head()
```

```python
df_boston['reg'] = list(map(lambda i:'reg_{0}'.format(i+1), range(df_boston.shape[0])))
df_boston.head()
```

```python
features_columns = set(df_boston.columns) - {'reg', 'target'}
columns_order = list(features_columns)
columns_order
```

```python
columns_order.insert(0, 'reg')
columns_order.insert(len(columns_order), 'target')
columns_order
```

```python
df_boston = df_boston[columns_order]
df_boston.head()
```

```python
df_boston.to_csv('boston_dataset.csv', sep=';', index=False)
```
