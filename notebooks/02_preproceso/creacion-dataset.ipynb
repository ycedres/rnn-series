# 2. - Selección de features

```python
def features_generator(ts, feature_window=10, horizont=1, padding=0):
    """
    Generator of features
    
    ts: array
    feature_window: number of points to split the ts
    horizont: ahead
    padding: ???
    """
    timesteps = len(ts) - (feature_window + horizont + padding - 1)
    
    features = np.zeros((timesteps, feature_window), dtype = np.float32)
    
    for t in range(padding, timesteps):
        # features
        features[t][0: feature_window] = ts[t : t+feature_window]
        
    return features


def labels_generator(ts, feature_window=10, horizont=1, padding=0):
    """
    Generatoe of label using ahead prediction strategy
    
    ts: array
    feature_window: number of points to split the ts
    horizont: ahead
    padding: ???
    """
    timesteps = len(ts) - (feature_window + horizont + padding - 1)
    
    labels = np.zeros((timesteps), dtype=np.float32)
    
    for t in range(padding, timesteps):
        # labels
        offset = t + feature_window + horizont - 1
        labels[t] = ts[offset]
        
    return labels
```

```python
def create_aheadPrediction_dataset(timeSeries,
                                   window_size=10,
                                   H=24,
                                   padding=0):
    """
    Create a list of dataframe. Each DF is ML dataset with 
    number of features same of window_size, and label the 
    ahead value.
    
    timeSeries: ts array
    window_size: number of points to split the ts
    H: horizon, range(len(H)) will be all possible horizon
    padding: 
    """
    # lista vacia
    dataSets = []
    
    for h in range(1, H+1):
        features = features_generator(timeSeries, feature_window=window_size, 
                                      horizont=h, padding=0)
        labels = labels_generator(timeSeries, feature_window=window_size, 
                                  horizont=h, padding=0)
    
        df = pd.DataFrame(features)
    
        df['h_{0}'.format(h)] = labels
    
        dataSets.append(df)
        
    # if toCSV == True: from DF to CSV
        
    return dataSets
```

# 4.- Volcado de los dataset a csv

```python
def from_pandas_to_csv(df, pathFileName):
    """
    
    df: df which wants to CSV file form.
    pathFileName: path 
    """
    # se incluye el tipo de formato: CSV
    pathFileName = '{0}.csv'.format(pathFileName)
    
    # se ejecuta: from DF to CSV
    df.to_csv(pathFileName, sep=';',
              index=False)
    
    print('Written file at {0}'.format(pathFileName))
```
