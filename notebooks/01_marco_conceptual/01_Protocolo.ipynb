# Protocolo para el análisis de Series Temporales

Este documento presenta los diferentes pasos de actuación para el análisis y predicción de series temporales con modelos de Machine Learning (ML) y Redes Neuronales Profundas (RNP).


## 1) Cargar los datos

Formatos:
 
 * CSV
 * Web
 * Picked
 * BBDD Relacionales
 * HDF5


## 2) Selección del *lag* o ventana

 * 2.1) **Soft-models**

  Se proponen un modelo *soft* del tipo *rolloig mean* con diferentes frecuencias y se comparan cada uno de ellas con el modelo original. 
        
  El modelo *soft* con mejor coeficiente de Pearson, su frecuencia se seleccionará como futura ventana para el entrenamiento de los modelo de ML y RNP:
  
  
 * 2.2) **Modelo autoregresivo**
 
 Para utilizar el modelo autoregresivo es necesario que el modelo sea *estacioanario* (media y desviación estandar constante a lo largo del tiempo). Si la serie no lo es habrá que ir suavizando su tendencia, estacionalidad y su ruido hasta que su media y desviación estandar sea lo más constante en el tiempo.
 
 La seria ahora estacionaria se le aplica un modelo autoregresivo (AR) proponiendo diferentes ordenes a la misma. Cada orden representa un conjunto de valores de la serie, equivalente a una ventana.
 
 Combinando este método con una métrica de error (por ejemplo: AIC) se puede obtener el mejor modelo para diferentes ordenes de AR propuesto. Con el mejor de ellos, su orden su ulizar para definir el *lag* o tamaño de la ventana.
 

## 3 ) Generación de "features"

Transformamos la serie en un espacio de *features - label* con el método de la ventana (el tamaño de dicha ventana ya se abría definido en el punto 2)

La serie temporal ahora quedería representada por:

 * features: número de puntos de la serie que posee la ventana.
 * label: es valor de la serie que se quiere predecir (*).
 * examples: número de medidas ( número de puntos de la serie - tamaño de la ventana - 1).
 
(*) Existen diferenres estrategías para calcular el valor de la serie que se quiere medir,  estas son enumeran:

 * predicción one-step ahead
 
 * predicción multi-step ahead:
  
    * iterativos
    * iteraticos donde los parámetros son optimizados para minimizar el error de entrenamiento sobre la predicción one-step-ahead.
    * Directa
    * Direc
    * MIMO


## 4 ) Separamos el conjunto de datos 

 * train: 80 % | 80 %  | 64 %
 * valid:      | 20 %  | 16 %
 * test:  20 %         | 20 %  
 
 
 
 
## 5 ) Dependiendo del *modelo* normalizamos los datos o no.

 * 5.1) Modelos con datos no normalizados: no se hace nada.
 
 * 5.2) Modelo con datos normalizados (RNP): 
 
   **Normalizamos**:
   
   train --> trainN
   
   val   --> valN
   
   test  --> testN
   
   * 5.2.1) RNP
   
     (trainN, valN) --> entrenamos el modelo y medimos el over-underfitting.
     
     (test) --> predN
     
     
## 6) Desnormalizamos los datos

predN --> pred

Se obtine finalmente: **( test, pred )**.
